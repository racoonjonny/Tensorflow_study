```python
# Image Data Augmentation이란
# CNN 모델의 성능을 높일 수 있는 방법 중 하나로써 원본 이미지에 적절한 변형을 가해서 새로운 데이터를 만들어내는 방식

# Keras 에서는 친절하게도 ImageDataGenerator라는 라이브러리를 사용할 수 있다.
# flow(), flow_from_directory() 등의 함수를 통해 이미지 데이터를 보강할 수 있음.
```

```python
# 1. flow() 를 이용한 방법
# ImageDataGenerator 생성
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.preprocessing.image import ImageDataGenerator

gen = ImageDataGenerator(rotation_range = 30, width_shift_range = 0.3, shear_range = 0.4, horizontal_flip = True)

# 이미지 load 및 정규화
image_array_list = [] 
img_names = ['./dog1.jpg', './dog2.jpg', './cat1.jpg', './cat2.jpg']

for i in range(len(img_names)):
  loaded_img = load_img(img_names[i], target_size = (100,100))

  loaded_img_array =  img_to_array(loaded_img) / 255.0 # normalization

  image_array_list.append(loaded_img_array)

plt.figure(figsize = (6,6))
for i in range(len(image_array_list)):
  plt.subplot(1, len(image_array_list), i+1)
  plt.xticks([]); plt.yticks([])
  plt.title(img_names[i])
  plt.imshow(image_array_list[i])
```

```python
batch_siz = 2
data_gen = gen.flow(np.array(image_array_list), batch_size= batch_siz)

img = data_gen.next() # batch_siz = 2 지정된 개수만큼 이미지 생성됨.

plt.figure(figsize= (6,6))
for i in range(len(img)):
  plt.subplot(1, len(img), i+1)
  plt.xticks([]); plt.yticks([])
  plt.imshow(img[i])
  
# 원본 이미지에서 변형된 결과가 생성됨을 확인할 수 있다.
# Keras 라이브러리를 사용하면 이러한 이미지 변형을 아주 쉽게할 수 있음.
```
